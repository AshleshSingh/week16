## Push to GitHub
Create an empty GitHub repository first. Then run the next cell. Use a GitHub Personal Access Token when prompted.

````python
from getpass import getpass
GITHUB_USERNAME = input('GitHub username: ')
GITHUB_EMAIL = input('GitHub email: ')
REPO_NAME = input('GitHub repo name: ')
TOKEN = getpass('GitHub Personal Access Token: ')

!git config --global user.name "{GITHUB_USERNAME}"
!git config --global user.email "{GITHUB_EMAIL}"
!git init
!git add .
!git commit -m "Initial AI Scheduler OS prototype"
!git branch -M main
!git remote remove origin 2>/dev/null || true
!git remote add origin https://{GITHUB_USERNAME}:{TOKEN}@github.com/{GITHUB_USERNAME}/{REPO_NAME}.git
!git push -u origin main
````
